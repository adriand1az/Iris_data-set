# Iris Flower Set
This data set consists of 149 samples from each of three species of the Iris flower. Im not sure why but there is a missing value for the Iris-setosa set. The length and the width of the sepals and petals, were measured in centimeters. My goal is to find if there are any simalarities between these three species

```python
#  Loading in imports
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#  Loading in data
df = pd.read_csv('iris.csv')

#  The data set didnt have any column headers so I created some
df.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "iris_type"]

```

## Examine the data

```python
#  Glimpse of the data
df.head()
```

```python
df.groupby(df['iris_type']).count()

#  We are missing a value for 'Iris-setosa' 
#  However a slight imperfection shouldnt be an issue
```

```python
#  Break each species into their own group
setosa = df.loc[df['iris_type'] == 'Iris-setosa']
virginica = df.loc[df['iris_type'] == 'Iris-virginica']
versicolor = df.loc[df['iris_type'] == 'Iris-versicolor' ]

#  Overview of data for each species
df.groupby('iris_type').mean()
```

# Description Of Each Type

# Data Visualization

```python
#  Box plot
df.boxplot(by='iris_type', figsize=(15,10))
```

# Sepal Length

```python
plt.figure(figsize=(10,4))
plt.title('Sepal length')
plt.xlabel('Species')
plt.ylabel('Measurement in (cm)')
plt.bar(df['iris_type'], df['sepal_length'])
```

# Sepal Width

```python
plt.figure(figsize=(10,4))
plt.title('Sepal Width')
plt.xlabel('Species')
plt.ylabel('Measurement in (cm)')
plt.bar(df['iris_type'], df['sepal_width'])
```

# Petal Length

```python
plt.figure(figsize=(10,4))
plt.title('petal length')
plt.xlabel('Species')
plt.ylabel('Measurement in (cm)')
plt.bar(df['iris_type'], df['petal_length'])
```

# Petal Width

```python
plt.figure(figsize=(10,4))
plt.title('Petal Width')
plt.xlabel('Species')
plt.ylabel('Measurement in (cm)')
plt.bar(df['iris_type'], df['petal_width'])
```

```python
plt.figure(figsize=(14,6)) # Setting size for the graph

plt.title('Stats') # Setting title for graph

sns.lineplot(data= df.iloc[:, 0:4]) #  Plotting the data

```

# Line graph

```python
# Set the width and height of the figure
plt.figure(figsize=(10,4))

# Add title
plt.title("Data")

sns.lineplot(data=setosa['petal_length'], label='Iris-setosa')
sns.lineplot(data=versicolor['petal_length'], label='Iris-versicolor')
sns.lineplot(data=virginica['petal_length'], label='Iris-virginica')

```

# Findings
Petal width and petal length seem to be proportionate through out all three species
* Virginica seems to be the largest all around flower
